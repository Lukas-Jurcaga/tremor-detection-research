```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
final_dataset_path = '../../dataset/original/'
```

```python
def get_all_raw_data(paths, axis):
    all_raw_data = []
    for path in paths:  
        # Read the CSV file and extract the desired axis
        data = pd.read_csv(final_dataset_path + path)[axis]
        
        # Standardize the dataframe using Z-score normalization
        data_standardized = (data - data.mean()) / data.std()
        
        # Append the standardized dataframe to a list
        all_raw_data.append(data_standardized.to_numpy())
    return all_raw_data
```

```python
def plot_hist(x_acc_data, y_acc_data, z_acc_data):
    plt.hist(x_acc_data, bins=30, density=True, alpha=0.6, color='g', label='X Acceleration')
    plt.hist(y_acc_data, bins=30, density=True, alpha=0.6, color='r', label='Y Acceleration')
    plt.hist(z_acc_data, bins=30, density=True, alpha=0.6, color='b', label='Z Acceleration')
    plt.xlabel('Acceleration (g)')
    plt.ylabel('Density')
    plt.legend()
    plt.title('Normalized Histogram of Accelerometer Data')
```

```python
paths = ['Abdullah/20231206_114621_Acceleration_processed.csv', 'Encarna/20231206_113808_Acceleration_processed.csv', 'Faezeh/20231206_112903_Acceleration_processed.csv', 'Marta/20231208_150416_Acceleration_processed.csv', 'Sadeque/20231206_112138_Acceleration_processed.csv', 'Tiantao/20231206_115916_Acceleration_processed.csv']
 

all_raw_x_acc_data = get_all_raw_data(paths, 'X')
all_raw_y_acc_data = get_all_raw_data(paths, 'Y')
all_raw_z_acc_data = get_all_raw_data(paths, 'Z')
```
